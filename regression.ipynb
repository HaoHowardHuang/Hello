```python
import pandas as pd
import statsmodels.api as sm 
```

```python
testdf = pd.DataFrame({'x':[1,2,3,4],'y':[10,200,50,100],'y1':[10,20,30,40]})
testmd = sm.OLS(testdf['x'],testdf[['y','y1']]).fit()
testmd.summary()
testmd.pvalues[0]
```

```python
df=pd.read_excel('TRD_Mnth.xlsx',sheet_name='Sheet1',converters={'Stkcd':str})
beta1 = []
year = []
company = []
a = float(1)
NaN = float('nan')
for gname1,glist1 in df.groupby('Stkcd'):
    for gname2,glist2 in glist1.groupby('year'):
        model = sm.OLS(glist2['Rs-Rn'],glist2[['Exreturn','RiskPremium1','SMB1','HML1']]).fit()
        if model.pvalues[0] is NaN:
            continue
        print(model.pvalues)
        print(model.params)
        if float(model.pvalues[0])<float(a):
            beta1.append(model.params[0])
            year.append(gname2)
            company.append(gname1)
```
